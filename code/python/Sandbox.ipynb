```python
from nltk.corpus import abc
tokens = ["one", "two", "three", "one", "two", "three", "one", "two", "three", "one", "two", "three",
         "four", "five", "six", "seven", "one", "two", "three", "one", "two", "three", "eight", "nine", 
         "one", "two", "three", "one", "two", "three", "one", "two", "three", "one", "two", "three",
         "four", "five", "six", "seven", "one", "two", "three", "one", "two", "three", "eight", "nine"
         ]


s = set(tokens)
freqs = []
for t in s:
    freqs.append((t, tokens.count(t)))

freqs.sort(key = lambda x: x[1])
    
thresh = 30
if len(freqs) < 30:
    thresh = len(freqs)
t30 = 0
for i in range(thresh):
    t30 += freqs[i][1]
print(f'Top {thresh}: {t30}, 30% {len(tokens) * 0.3}') 
```

```python
import zipfile
import os

input_file = zipfile.ZipFile(f'subtitles-en.zip', 'r')
output_file = zipfile.ZipFile(f'subtitles-en-small.zip', 'a')

years = set()


for f in input_file.namelist():
    if f.endswith('xml'):
        if f.startswith('OpenSubtitles/raw/en'):
            year = int(f.split('/')[3])
            if year not in years:
                years.add(year)
                output_file.writestr(f, input_file.open(f).read())
output_file.close()
```

```python
import re
from lxml import etree

expressions = [
    (r'<.*?>', ''),  # xml tags
    (r'http.*?(?:[\s\n\]]|$)', ''),  # links
    (r'\s\(.*?\)', ''),  # parentheses
    (r'([^\s]{2,})[\.\!\?\:\;]+?[\s\n]|$', '\\1\n'),  # break sentences at periods
    (r"[-–—/']", ' '),  # hyphens, apostrophes and slashes
    (r'\s*\n\s*', '\n'),  # empty lines
    (r'\s{2,}', ' '),  # excessive spaces
]
patterns = [(re.compile(expression[0], re.IGNORECASE), expression[1]) for expression in expressions]

def clean_text(text):
    txt = text
    for pattern in patterns:
        txt = pattern[0].sub(pattern[1], txt)
    txt = ''.join([letter for letter in txt if (letter.isalnum() or letter.isspace())]) 
    return txt

def sub_xml_to_text(xml, parser):
    tree = etree.fromstring(xml, parser)
    for node in tree.iter():
        if node.tag == 'meta':
            tree.remove(node)
    return etree.tostring(tree, encoding=str, method='text')

xml_stuff = """
<stuff>
  <nonsense>
    Things.  Things? (Really?)  Yes, things!
  </nonsense>
  <nonsense>
    Things.  Things? http://things.com Yes, things!
  </nonsense>
  <nonsense>
    Things.  Things?  Things-which are thing's things and/or things. Yes, things!
  </nonsense>
  <nonsense>
    Things.  Things?  Yes, things!
  </nonsense>
  <nonsense>
    Things.  Things?  Yes, things!
  </nonsense>
  <meta>
    <incomprehensible>
    Jabber.
    </incomprehensible>
  </meta>
</stuff>
"""
xmlparser = etree.XMLParser(recover=True, encoding='utf-8')
text = sub_xml_to_text(xml_stuff, xmlparser)
new_text = clean_text(text)
print(new_text)
```

```python
import bz2
def wiki_articles(language):
    myfile = bz2.open(f'wikipedia-{language}.bz2', 'rt', encoding='utf-8')
    
    article = ""
    body = False
    
    line = myfile.readline()
    
    while line:
        if "<page>" in line:
            print("Page")
            body = True
        
        if "</page>" in line:
            body = False
            print("End Page")
            
        if body:
            print("Body")
            
        
        line = myfile.readline()
        
    myfile.close()
```
