```python
import glob
import matplotlib.pyplot as plt
import matplotlib
import matplotlib as mpl
import numpy as np
import os
import pandas as pd
import sys
from IPython.display import display
```

```python
basedir = os.path.join('Y:\\', 'Dissertation')
languages = set()
agg = list()

# Start with replication
rep = os.path.join('Z:\\', 'Dissertation', "scores", "top_3_by_language.csv")
df = pd.read_csv(rep, header=0)
for index, stuff in df.iterrows():
        agg.append(stuff[['Language','Dimensions','Window','Algorithm','Score']])
        languages.add(stuff['Language'])
        
# Add frequencies
rep = os.path.join(basedir, "count_evals", "subs_top_3_per_algo.csv")
df = pd.read_csv(rep, header=0)
for index, stuff in df.iterrows():
        agg.append(stuff[['Language','Dimensions','Window','Algorithm','Score']])
        languages.add(stuff['Language'])
        
rep = os.path.join(basedir, "count_evals", "wiki_top_3_per_algo.csv")
df = pd.read_csv(rep, header=0)
for index, stuff in df.iterrows():
        agg.append(stuff[['Language','Dimensions','Window','Algorithm','Score']])
        languages.add(stuff['Language'])
        
# Finally the additional datasets
rep = os.path.join(basedir, "scores", "top_3_per_algo.csv")
df = pd.read_csv(rep, header=0)
for index, stuff in df.iterrows():
        agg.append(stuff[['Language','Dimensions','Window','Algorithm','Score']])
        languages.add(stuff['Language'])
```

```python
agg_df = pd.DataFrame(agg)
retval = list()
pd.options.mode.copy_on_write = True
for lang in languages:
    df_lang = agg_df.loc[agg_df['Language'] == lang]
    df_lang.sort_values(by=['Score',"Dimensions","Window"], ascending=[False, True, True], inplace=True)
    retval.append(df_lang.iloc[0])
    
retval_df = pd.DataFrame(retval)
retval_df.sort_values(by=['Language'], ascending=[True], inplace=True)

# Fix up column values
code2lang = os.path.join(basedir, "geo", "code2lang.csv")
code_df = pd.read_csv(code2lang)
retval_df.columns = ['code', 'Dimensions', 'Window', 'Algorithm', 'Score']
retval_df = retval_df.merge(code_df, on="code",how='left')
retval_df.loc[retval_df["Algorithm"] == "cbow", "Algorithm"] = 'CBOW'
retval_df.loc[retval_df["Algorithm"] == "sg", "Algorithm"] = 'Skip-gram'
retval_df = retval_df.drop(columns=['code'])
retval_df = retval_df[['Language', 'Dimensions', 'Window', 'Algorithm', 'Score']]
```
